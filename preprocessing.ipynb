```python
import pandas as pd
import spacy, nltk, gensim, sklearn
```

```python
df = pd.read_csv('raw_input.csv', index_col= 0)
df.head()
```

```python
#Removing preffix and suffix from company name
df['Company'] = df['Company'].apply(lambda x: x[8:-9])
#For some reason some duplicates in the CSV
df = df.drop_duplicates()
df.head()
```

```python
nlp1 = spacy.load('en_core_web_sm', exclude=['parser'])
nlp1.enable_pipe("senter")
nlp2 = spacy.load('en_core_web_sm')
```

```python
#Function to get list of sentences from one report
def get_list1(company):
    df_company = df[df['Company'] == company]
    text = df_company['text'][0]
    doc = nlp1(text)
    sentences = [sent for sent in doc.sents]
    return sentences

def get_list2(company):
    df_company = df[df['Company'] == company]
    text = df_company['text'][0]
    doc = nlp2(text)
    sentences = [sent for sent in doc.sents]
    return sentences
    
    
```

```python
Tesla1 = get_list1('Tesla')
Tesla2 = get_list2('Tesla')
Tesla1
```
