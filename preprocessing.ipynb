```python
import pandas as pd
import spacy, nltk, gensim, sklearn
```

```python
df = pd.read_csv('raw_input.csv', index_col= 0)
df.head()
```

```python
#Removing preffix and suffix from company name
df['Company'] = df['Company'].apply(lambda x: x[8:-9])
#For some reason some duplicates in the CSV
df = df.drop_duplicates()
df.head()
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
#Function to get list of sentences from one report
def get_list(company):
    df_company = df[df['Company'] == company]
    text = df_company['text'][0]
    doc = nlp(text)
    sentences = [sent for sent in doc.sents]
    return sentences
    
```

```python
Tesla = get_list('Tesla')
Tesla
```

```python
Amazon = get_list('Amazon')
```

```python
# Printing random sentences to check if it works
sentences = [sent for sent in Amazon]
print('Sentence 1:',sentences[1],'\n')
print('Sentence 2:',sentences[0],'\n')
print('Sentence 3:',sentences[501],'\n')
```
